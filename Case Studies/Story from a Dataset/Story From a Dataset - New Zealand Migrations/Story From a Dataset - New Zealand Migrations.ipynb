## Arrivals

```python
arrivals = nz[nz['Measure'] == 'Arrivals']
```

#### Immigration to new zealand over the years

```python
# Arbitrarily choosing to focus on rows marked with citizenship as "Total All Citizenships"
immigration = arrivals[arrivals['Citizenship'] == 'Total All Citizenships']

# Grouping immigration by year and taking the sum of immigrations per year
immigration = immigration.groupby('Year').sum().reset_index().rename(columns={'Value':'Migration Volume'})
```

```python
sns.regplot(data=immigration, x='Year', y='Migration Volume', truncate=False)
plt.title('Immigration to New Zealand Per Year');
```

We can see that immigration to New Zealand has stayed relatively steady at a flow of about 6,000 people per year while increasing sharply in recent years.

## Departures

```python
departures = nz[nz['Measure'] == 'Departures']
```

#### Emigration from new zealand over the years

```python
# Arbitrarily choosing to focus on rows marked with citizenship as "Total All Citizenships"
emigration = departures[departures['Citizenship'] == 'Total All Citizenships']

# Grouping immigration by year and taking the sum of immigrations per year
emigration = emigration.groupby('Year').sum().reset_index().rename(columns={'Value':'Migration Volume'})
```

```python
sns.regplot(data=emigration, x='Year', y='Migration Volume', truncate=False)
plt.title('Emigration to New Zealand Per Year');
```

We can see that emigration from New Zealand has been more volatile, fluctuating around a flow of 1,600 people per year, decreasing sharply recently.

## Net migration to New Zealand per country

#### Top 10 countries by net migration volume over the years

```python
# Drop rows where country isn't a country, found these by "fishing" for them
not_countries = ('All countries','Africa and the Middle East','Asia','Europe','Antarctica','Not stated','Americas','Oceania')
net_real_countries = net[~net['Country'].isin(not_countries)]
```

```python
# Find top 10 countries by sum of net migration values, then save their country names
top10 = net_real_countries.groupby('Country')[['Value']].sum('Value').nlargest(10,'Value').index.tolist()

# Using the top 10 countries, query the net dataframe for all rows containing any of the countries
top10 = net[net['Country'].isin(top10)]
```

```python
# Pivot the top 10 migrating countries' data so columns are countries, rows are years, and values are the net volume
# of migration per country
net_migration_per_country = pd.pivot_table(top10, values='Value', index='Year', columns='Country')
```

```python
net_migration_per_country.plot(figsize=(15,8), title='Top 10 Countries by Net Migration Volume', ylabel='Net to/from NZ');
```

Within the top 10 migrating countries it's interesting to see how countries that are relatively close to New Zealand are the top ones immigrating and staying by net volume, however, these are also some of the most populous countries in the world.

Some outliers amongst this list are Germany, South Africa, and the UK. I would hypothesize that since New Zealand was a colony it's easier for South Africans and UK people to travel there (as in, culturally).

It's also of special notice that the UK has had some of the most volatile net volume of migration to New Zealand, with sharp swings one way or the other over the years.

### Create top/bottom countries by net migration per year function

```python
def gen_top_bottom_migration_per_year(year, n):
    """
    Generate the top and bottom n migrating countries per year as a dataframe and as a barplot of the data.
    Returns a dataframe and a matplotlib barh plot.
    """
    
    year = net_real_countries[net_real_countries['Year'] == year].sort_values('Value', ascending=False)
    
    top = year[:n]
    bottom = year[-n:]
    top_and_bottom = pd.concat([top, bottom])
    
    y = top_and_bottom['Country']
    width = top_and_bottom['Value']
    
    plt.barh(y=y, width=width)
    
    return top_and_bottom
```
