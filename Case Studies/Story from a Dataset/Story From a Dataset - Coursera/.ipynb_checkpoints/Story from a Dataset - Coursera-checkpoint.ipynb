# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import Data

```python
df = pd.read_csv('coursera_data.csv')
```

```python
df.head()
```

# Data Cleaning

```python
# Rename columns
df.columns = ['dataset','title','organization','certification_type','rating','difficulty','students_enrolled']
df.head()
```

```python
# Check the feature datatypes
df.dtypes
```

```python
# Correct inappropriate datatype
df.students_enrolled = (df.students_enrolled.str.replace('k','').str.replace('m','').astype(float) * 1000).astype(int)
```

```python
df.dtypes
```

# Exploratory Data Analysis

```python
sns.set_theme(context='notebook', style='darkgrid')
```

## Number of Courses

```python
data = df['organization'].value_counts().reset_index()[:40]
data.columns = ['Organization', 'Counts']

plt.figure(figsize=[8,8])
sns.histplot(data=data, y='Organization', weights='Counts')
plt.title('Top 40 Coursera Organizations by # of Courses');
```

## Advanced Difficulty Courses

```python
advanced = df[df['difficulty'] == 'Advanced']
advanced_sortedByStudents_desc = advanced.sort_values('students_enrolled', ascending=False)
advanced_sortedByRating_desc =advanced.sort_values('rating', ascending=False)
```

```python
sns.histplot(data=advanced_sortedByStudents_desc[:10], weights='students_enrolled', y='title', \
             hue='certification_type', alpha=1)
plt.title('Top Advanced Courses by Sum of Enrolled Students ')
plt.xlabel('Sum of Enrolled Students')
plt.ylabel('Course Title');
```

```python
sns.histplot(data=advanced_sortedByRating_desc[:10], weights='students_enrolled', y='title', \
             hue='certification_type', alpha=1)
plt.title('Top Advanced Courses by Rating')
plt.xlabel('Sum of Enrolled Students')
plt.ylabel('Course Title');
```

## University of Pennsylvania Data

```python
upenn_data = df[df['organization'] =='University of Pennsylvania']

upenn_top_students_enrolled = upenn_data.sort_values('students_enrolled', ascending=False)[:5]

upenn_data.sort_values('rating', ascending=False)[:5]
```

```python
# Aggregate numeric dtypes
top_orgs_by_sum_students = df.groupby('organization').agg({'rating':'median', 'students_enrolled':'sum'}, axis=1)

# Find the most common ___ per organization
top_orgs_by_sum_students['certification_type'] = df.groupby('organization')['certification_type'].agg(pd.Series.mode).astype(str)
top_orgs_by_sum_students['difficulty'] = df.groupby('organization')['difficulty'].agg(pd.Series.mode)

# Sort the dataframe by sum of students enrolled and reset the index to be a feature
top_orgs_by_sum_students = top_orgs_by_sum_students.sort_values('students_enrolled', ascending=False).reset_index()

# Rename features
top_orgs_by_sum_students.columns = ['organization','MEDIAN_rating','SUM_students_enrolled', \
                                    'MODE_certification_type','MODE_difficulty']

# Print the first 5 rows
top_orgs_by_sum_students.head()
```

## Organizations

```python
sns.histplot(data=top_orgs_by_sum_students[:20], y='organization', weights='SUM_students_enrolled', \
             hue='MODE_certification_type', alpha=1)
plt.title('Top 20 Coursera Organizations by Sum of Enrolled Students');
```

```python
sns.histplot(data=top_orgs_by_sum_students[:10], weights='SUM_students_enrolled', \
             y='organization', hue='MEDIAN_rating', alpha=1);
```

```python
plt.figure(figsize=[12,8])
sns.scatterplot(data=df[:60], x='rating', y='organization', size='students_enrolled', hue='difficulty')
plt.title('Students Enrolled vs. Rating');
```
