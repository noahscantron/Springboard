### Task 1:
#### Collect Alibaba stock (HKEX 09988) data for all of 2020

```python
# Configure parameters to pass to Quandl API to return desired dataset as "payload"
payload = {'start_date':'2020-01-01','end_date':'2020-12-31','order':'asc','api_key':API_KEY}
url = f'https://www.quandl.com/api/v3/datasets/HKEX/09988.json?'
r = requests.get(url,params=payload)
```

```python
# Check if the API request worked
r
```

### Task 2:
#### Convert the returned JSON to a Python dictionary

```python
alibaba = dict(r.json())
```

```python
# Proof that the returned JSON has been converted to a Python dictionary via access example
alibaba['dataset']['id']
```

### Task 3:
#### Calculate the highest and lowest prices for 2020, I'll be calculating the highest and lowest intraday price as 'open' isn't a feature of this dataset

```python
alibaba.keys()
```

```python
alibaba['dataset'].keys()
```

```python
column_names = alibaba['dataset']['column_names']
column_names
```

```python
# Check the first two stock records in the dataset's records
data = alibaba['dataset']['data']
data[:2]
```

```python
# Use a for-loop to compare each record's high and low to the previous high and lows, replacing them if greater or smaller
maximum = float('-inf')
minimum = float('inf')

for x in data:
    high = x[7]
    low = x[8]
    maximum = max(maximum,high)
    minimum = min(minimum, low)
```

```python
# The highest opening price
print(f'The highest opening price in 2020 was {maximum} Hong Kong Dollars')
```

```python
# The lowest opening price
print(f'The lowest opening price in 2020 was {minimum} Hong Kong Dollars')
```

### Task 4:
#### What was the largest change in any one day (based on high/low prices)?

```python
max_chg = float('-inf')

for x in data:
    high = x[7]
    low = x[8]
    max_chg = max(max_chg, high - low)
```

```python
# The largest change
print(f'The largest change in any one trading day was {max_chg} Hong Kong Dollars')
```

### Task 5:
#### What was the largest change between any two days (based on closing-price)?

```python
max_closing_chg = float('-inf')
prev_close = data[0][9]

for x in data:
    close = x[9]
    chg = close - prev_close
    max_closing_chg = max(max_closing_chg, chg)
```

```python
# The largest change between any adjacent closing-prices
print(f'The largest closing price change was {max_closing_chg} Hong Kong Dollars')
```

### Task 6:
#### What was the average daily trading volume this year?

```python
sum = 0
count = 0

for x in data:
    count += 1
    sum += x[10]
```

```python
avg_volume = sum / count
```

```python
# The average daily trading volume
print(f'The average daily trading volume was {round(avg_volume)} trades per day')
```

### Task 7:
#### Find the median trading volume for 2020

```python
ordered_volumes = [x[10] for x in data]
ordered_volumes.sort()
```

```python
median = ordered_volumes[len(ordered_volumes) // 2]
```

```python
# The median trading volume for 2020
print(f'The median trading volume for 2020 was {round(median)} trades per day')
```
